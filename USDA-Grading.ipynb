The USDA Agricultural Marketing Service provides voluntary beef grading services according to established official USDA beef quality and yield grade standards. Beef grades are based on two primary criteria: quality grade, which is largely influenced by the degree of marbling and maturity, and yield grade, which represents the carcass cutability.

1. **Quality Grading**: A measure of the animal's age and the amount of marbling in the meat. Marbling, the intramuscular fat distributed within the muscle, is a critical factor determining the palatability of beef (flavor, juiciness, and tenderness). Age is another factor considered in quality grading, as younger animals tend to produce more tender meat.

   Quality grades, listed from highest to lowest quality, include:

   - Prime: Comes from young, well-fed beef cattle. It has abundant marbling and is generally sold in restaurants and hotels.
   - Choice: High quality, but has less marbling than Prime. It is the most widely available grade in the market.
   - Select: Very uniform in quality and leaner than the higher grades. It is fairly tender but, because it has less marbling, it may lack some of the juiciness and flavor of the higher grades.
   - Standard and Commercial: Frequently sold as ungraded or as "store brand" meat.
   - Utility, Cutter, and Canner: Mostly used for ground beef and processed products.

2. **Yield Grading**: This is a measure of the proportion of the carcass in the lean cuts (the round, loin, rib, and chuck). The yield grade is represented on a scale from 1 to 5. A Yield Grade 1 carcass provides the highest yield of lean meat, while a Yield Grade 5 will have the lowest yield of lean meat.

The USDA grader determines the yield grade of a carcass by evaluating:

   - The amount of external fat.
   - The amount of kidney, pelvic, and heart fat.
   - The area of the ribeye muscle.
   - The hot carcass weight.

The final yield grade is calculated using a mathematical formula that takes these factors into account.

It's important to note that while USDA grading is a voluntary service provided by the U.S. government, it is widely used and sets the standard for beef quality in the U.S. 

To summarize, USDA beef grading practices involve a comprehensive evaluation of both the quality and quantity (yield) of beef produced from a carcass. The goal of these practices is to provide producers and consumers with a reliable system for assessing and communicating beef quality.

**Yield Grades**: This measures the amount of usable lean meat on the carcass. Yield grades range from 1 to 5, with 1 being the highest yield (most lean meat) and 5 being the lowest yield (least lean meat).

The official grading is performed by a USDA grader who has been highly trained to evaluate beef carcasses for these factors.

However, the USDA does not use a numerical scoring system from 100 to 1100 for beef grading. If such a scoring system exists, it might be specific to a certain company, research project, or specialized grading program outside the official USDA grading system.
