```python
loc="./../../data/marble/iteration-1/choice/test"
i=0

for path, subdirs, files in os.walk(loc):   ### Walk sub directories and files
    for name in files:
        #print(os.path.join(path, name))
        fname = os.path.join(path, name)
        image = cv2.imread(filename = fname)
        image = image[0:500, 150:800]
        # create binary image
        gray = cv2.cvtColor(src = image, code = cv2.COLOR_BGR2GRAY)
        hsv = cv2.cvtColor(src = image, code = cv2.COLOR_BGR2HSV)
        # smooth the image to remove unwanted noise/unwanted details#
        blur = cv2.GaussianBlur(src = gray, ksize = (5, 5), sigmaX = 0)
        t=100
        # apply simple thresholding to convert the grayscale image to binary image
        (t, binary) = cv2.threshold(src = blur, thresh = t, maxval = 255, type = cv2.THRESH_BINARY)
        edges = cv2.Canny(gray,30,200)
        # define a (3, 3) structuring element
        kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
        # apply the dilation operation to the edged image
        white_lower = (0,0,140)
        white_upper = (255,255,255)
        mask_white = cv2.inRange(hsv, white_lower, white_upper)
        contours, _ = cv2.findContours(mask_white, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        #im = np.copy(image)
        #cv2.drawContours(im, contours, -1, (0, 255, 255), 2)
        count = 0
        for (i, c) in enumerate(contours):
            if len(c) > 500: continue
            count += 1

        #count += len(contours)
        print(path +'/'+ name, count)
```
